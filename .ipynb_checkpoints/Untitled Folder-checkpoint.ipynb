```python
# now I use vectorisation concept to detrmine reltion by 
#vector distance between tags
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words= 'english')
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return" ".join(y)
```

```python
#stem library example
['loved', 'love', 'loving']
ps.stem('loving')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
# with the help of this code we fetch five albums which is very similar
# our slected album 
sorted(list(enumerate(similarity[0])),reverse = True, key= lambda x:x[1])[1:6]
```

```python
#new_df[new_df['Album'] == 'Kid A'].index[0]
```

```python
def recommend(music):
    music_index = new_df[new_df['Album'] == music].index[0]
    distances = similarity[music_index]
    music_list = sorted(list(enumerate(distances)),reverse = True, key= lambda x:x[1])[1:6]
    
    for i in music_list:
        print(new_df.iloc[i[0]].Album)
```

```python
recommend('Call Me')
```
